```python
import pandas as pd
import matplotlib.pyplot as plt

# Create the DataFrame with three separate vectors for the horizontal axis
data1 = pd.DataFrame({
    'X1': [7,8,9,10,11,12],
    'Y': [20,30,40,50,60,70]
})

data2 = pd.DataFrame({
    'X2': [1,2,3,4,5,6],
    'Y': [12,13,16,18,20,22]
})


data3 = pd.DataFrame({
    'X3': [7,9,13,15,18,17],
    'Y': [19,7,11,11,15,18]
})




# Plotting the scatter plot
plt.scatter(data1['X1'], data1['Y'], color='yellow', label='Trend 1')
plt.scatter(data2['X2'], data2['Y'], color='orange', label='Trend 2')
plt.scatter(data3['X3'], data3['Y'], color='pink', label='Trend 3')

# Customize the plot
plt.xlabel('Horizontal Axis')
plt.ylabel('Vertical Axis')
plt.title('Scatter Plot with multiple Trends')
plt.legend()

# Show the plot
plt.show()
```

```python
#for loop
numbers = [10,15,20,25,30]
squared_numbers = []
for num in numbers:
    squared_numbers.append(num ** 3)
print(squared_numbers)
```

```python
#comprehension
numbers = [10,15,20,25,30]
squared_numbers = [num ** 3 for num in numbers] # [(operation on object) for object in iterable]
print(squared_numbers) 
```

```python
#for loop
numbers = [10,15,20,25,30]
even_numbers = []
for num in numbers:
    if num % 3 == 0:
        even_numbers.append(num)
print(even_numbers)  
```

```python
#comprehension
numbers = [10, 15, 20, 25, 30]
even_numbers = [num for num in numbers if num % 3 == 0]
print(even_numbers)
```

```python
data = {'A': [2,4,6,8,10,12], 'B': [14,16,18,20,22,24]}
df = pd.DataFrame(data)

# Define the lambda function to operate on multiple columns
function = lambda row: row['A'] * 2 + row['B']

# Apply the lambda function to create the new column
df['C'] = df.apply(function, axis=1)

# Display the updated DataFrame
print(df)
```

```python
import pandas as pd
import numpy as np

data = {'A': [2,4,6,8,10,12], 'B': [14,16,18,20,22,24]}
df = pd.DataFrame(data)

# Define the conditions
conditions = [
    df['A'] < 4,
    df['A'] >= 4
]

# Define the functions corresponding to each condition
def function1(x):
    return x * 1

def function2(x):
    return x + 2

# Vectorize the functions
vectorized_function1 = np.vectorize(function1)
vectorized_function2 = np.vectorize(function2)

# Apply the functions based on the conditions and create the new column
df['C'] = np.select(conditions, [vectorized_function1(df['A']), vectorized_function2(df['B'])], default=np.nan)

# Display the updated DataFrame
print(df)
```

```python
import pandas as pd
import numpy as np

data = {'A': [2,4,6,8,10,12], 'B': [14,16,18,20,22,24]}
df = pd.DataFrame(data)
df
# Define the conditions
conditions = [
    df['A'] < 20,
    df['A'] >= 16
]

# Define the functions corresponding to each condition
def function1(df):
    return df['A'] * 2 + df['B']

def function2(df):
    return df['A'] + df['B'] * 2

# Apply the functions based on the conditions and create the new column
df['C'] = np.select(conditions, [function1(df), function2(df)], default=np.nan)

# Display the updated DataFrame
print(df)
```
